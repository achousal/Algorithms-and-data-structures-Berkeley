```python
import random

#finds a random number in a random array in nlogn time

def Bsearch(arr, target):
    l , r = 0 , len(arr) - 1
    while l <= r:
        m = l + (r - l) // 2
        if arr[m] == target:
            return m
        elif arr[m] > target:
            r = m - 1
        else:
            l = m + 1
    return "not found"

def poparr(start=10, end=1000):
    arr = []
    n = random.randint(10, 1000)
    generated_numbers = set()  
    for i in range(n):
        rs = random.randint(start, end)
        while rs in generated_numbers:  
            rs = random.randint(start, end)
        generated_numbers.add(rs) 
        arr.append(rs)
    arr.sort()
    return arr

array = poparr()
print("Generated Array:", array)
target = random.choice(array)  # Selecting a random element from the array as target
print("Target:", target)
print("Index of Target:", Bsearch(array, target))
    
        
```
